```python
# PROGRAM: Two-Night Sleep Average Tracker

print("--- Welcome to the Sleep Tracker ---")

# Step 1: Start a safety net for errors
try:
    # Step 2: Get user name and sleep hours
    user_name = input("Enter your name: ")
    night1 = float(input("Enter hours slept on Night 1: "))
    night2 = float(input("Enter hours slept on Night 2: "))

    # Step 3: Math Expressions
    # Add the hours together first
    total_sleep = night1 + night2
    
    # Calculate the average (Division)
    average_hours = total_sleep / 2

    # Step 4: Multi-way Conditional Decisions
    print("\n--- Results for " + user_name + " ---")
    
    if average_hours >= 8:
        print("Status: Excellent. Your average sleep is perfect.")
    elif average_hours >= 6:
        print("Status: Fair. You are getting a decent amount of rest.")
    else:
        print("Status: Warning. Your average sleep is too low.")

    # Step 5: Final numeric summary
    print("Average Hours: " + str(average_hours))

except:
    # Step 6: This block runs if input is not a number
    print("Error: Please enter a number for your sleep hours.")

print("All Done")
```
