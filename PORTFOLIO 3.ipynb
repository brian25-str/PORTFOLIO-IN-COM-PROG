```python
# PROGRAM: Hydration Goal Tracker

# 1. A void function to greet the user (Standard task)
def welcome(name):
    print("Welcome, " + name + "!")
    print("Let's check your hydration progress.")

# 2. A function with two parameters (intake and goal)
def evaluate_hydration(current, target):
    if current >= target:
        print("Status: Success! You reached your goal.")
    else:
        # Math expression to find the missing amount
        missing = target - current
        print("Status: You need to drink " + str(missing) + " more liters of water today or for the week.")

# --- Main Program Execution ---
try:
    # 3. Built-in functions: input and float
    user_name = input("What is your name? ")
    welcome(user_name) # Calling a function with one argument

    my_goal = float(input("Enter your daily/weekly goal water intake (liters): "))
    my_intake = float(input("How much have you drunk so far (liters)? "))

    # 4. Calling the function with two arguments
    evaluate_hydration(my_intake, my_goal)

except:
    # 5. Error handling for non-numeric input
    print("Error: Please enter a number for goals and intake.")

print("All done!")
```
