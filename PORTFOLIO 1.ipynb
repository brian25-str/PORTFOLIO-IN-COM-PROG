```python
# PROGRAM: PHP Freelance Pay Calculator

# 1. Input and Type Conversion
# We convert input (string) to float to handle centavos (decimals)
hourly_rate_php = float(input("Enter your hourly rate in PHP: "))
hours_worked = float(input("Enter total hours worked: "))

# 2. Arithmetic Expressions
# Multiplication expression to find gross pay
gross_pay = hourly_rate_php * hours_worked

# Expression using precedence (Parentheses) to calculate 12% tax
# 0.12 is a constant value
tax_amount = gross_pay * (12 / 100)

# Subtraction expression to find the final take-home pay
net_pay = gross_pay - tax_amount

# 3. Output with String Concatenation
print("\n--- Payment Summary (PHP) ---")
print("Gross Pay: " + str(gross_pay))
print("Tax Deduction (12%): " + str(tax_amount))
print("------------------------------")
print("Total Net Pay in PHP: " + str(net_pay))
```
