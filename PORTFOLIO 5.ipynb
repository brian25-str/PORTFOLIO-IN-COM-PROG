```python
# PROGRAM: Simple Word Analyzer

# 1. Getting the input (always enters as a string)
raw_input = input("Enter a favorite word: ")

# 2. Cleaning the string
# .strip() removes spaces; .lower() makes it lowercase
word = raw_input.strip().lower()

# 3. Using len() to find the size
size = len(word)

if size > 0:
    # 4. String Indexing (Getting the first letter)
    first_letter = word[0]

    # 5. String Slicing (Getting the first 3 letters)
    # This means "start at 0, go up to but NOT including 3"
    short_version = word[0:3]

    print("\n--- Results ---")
    print("Cleaned Word: " + str.capitalize(word))
    print("Length: " + str(size) + " characters")
    print("First Letter: " + str.capitalize(first_letter))
    print("First 3 letters: " + str.capitalize(short_version))

    # 6. Using 'in' as a logical operator
    if 'a' in word:
        print("Search: The letter 'A' is in your word!")
    else:
        print("Search: The letter 'A' was not found :(")

else:
    print("Error: You didn't enter anything.")


```
