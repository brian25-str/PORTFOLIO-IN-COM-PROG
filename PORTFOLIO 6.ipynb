```python
# PROGRAM: Simple Task Collector

# 1. Create an empty list (Building from scratch)
task_list = list()

print("Build Your Task List")
print("Type your tasks one by one. Type 'done' when finished.")

# 2. Indefinite loop to collect items
while True:
    entry = input("Enter a task: ")
    
    # 3. Check for sentinel value to exit
    if entry.lower() == 'done':
        break
    
    # 4. Add the entry to the end of our collection
    task_list.append(entry)

# 5. Sorting the list alphabetically
task_list.sort()

# 6. Final Summary
print("\n--- Your Organized Tasks ---")
# Using len() to count items in the collection
print("Total Tasks: " + str(len(task_list)))

# 7. Definite loop to print each item
for task in task_list:
    print("- " + task)

print("\nAll tasks recorded!")
```
