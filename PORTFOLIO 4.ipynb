```python
# PROGRAM: Daily Workout Rep Tracker

# 1. Initialize variables for loop patterns
total_reps = 0
largest_set = None

print("5-Set Workout Logger")

# 2. Start a Definite Loop (runs exactly 5 times)

for set_num in [1, 2, 3, 4, 5]:
    try:
        # 3. Get input for the current set
        # Using concatenation to show the set number
        prompt = "Enter repetitions for exercise Set " + str(set_num) + ": "
        reps = int(input(prompt))

        # 4. Summing Pattern: Add to total
        total_reps = total_reps + reps

        # 5. Finding Largest Pattern: Track personal best
        if largest_set is None or reps > largest_set:
            largest_set = reps

    except:
        # Error handling if the user types words
        print("Invalid input. Please enter a whole number.")
        continue # Jump to the next set

# 6. Final Summary after the loop finishes
print("----------------------------")
if total_reps > 0:
    print("Total Repetitions: " + str(total_reps))
    print("Your Personal Best Set: " + str(largest_set))
    
    # Simple conditional logic to provide encouragement
    if total_reps >= 100:
        print("Status: Incredible work!")
    else:
        print("Status: Good effort. Keep pushing!")
else:
    print("No workout data recorded.")
```
